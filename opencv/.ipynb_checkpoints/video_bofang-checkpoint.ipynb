```python
#播放视频

import cv2

video_file=r'C:\Users\DFG\Videos\Captures\1.mp4'
cap=cv2.VideoCapture(video_file)

while True:
    ret,frame = cap.read()
    cv2.imshow('douying',frame)
    print('%.2f' % int(cap.get(0)/1000))
    if cv2.waitKey(5) == ord('q'):
        cv2.destroyAllWindows()
        cap.release()
    
cv2.destroyAllWindows()
cap.release()
```
